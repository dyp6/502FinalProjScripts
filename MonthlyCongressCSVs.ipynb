```python
import dask
import dask.dataframe as dd
import dask.bag as db
import json
```

```python
Months = ["01","02","03","04","05","06","07","08","09","10","11","12"]
dates17 = ["2017"+'-'+str(month) for month in Months[5:]]
dates18 = ["2018"+'-'+str(month) for month in Months]
dates19 = ["2019"+'-'+str(month) for month in Months]
```

```python
MonthlyJsons17 = list() 
for date in dates17:
    MonthlyJsons17.append(db.read_text('s3://502finalprojbucky/congresstweets/data/'+date+'*.json').map(json.loads))

MonthlyJsons18 = list()
for date in dates18:
    MonthlyJsons18.append(db.read_text('s3://502finalprojbucky/congresstweets/data/'+date+'*.json').map(json.loads))

MonthlyJsons19 = list()
for date in dates19:
    MonthlyJsons19.append(db.read_text('s3://502finalprojbucky/congresstweets/data/'+date+'*.json').map(json.loads))
```

```python
DFListMonth=list()
for days in MonthlyJsons17:
    DFListDay=list()
    for JsonList in days:
        b = db.from_sequence(JsonList)
        DFListDay.append(b.to_dataframe())
    DFListMonth.append(DFListDay)
```

```python
# June
CongDF_June17 = DFListMonth[0][0]
for DF in DFListMonth[0][1:]:
    CongDF_June17 = CongDF_June17.append(DF)
#July
CongDF_July17 = DFListMonth[1][0]
for DF in DFListMonth[1][1:]:
    CongDF_July17 = CongDF_July17.append(DF)
# August
CongDF_Aug17 = DFListMonth[2][0]
for DF in DFListMonth[2][1:]:
    CongDF_Aug17 = CongDF_Aug17.append(DF)
# September
CongDF_Sep17 = DFListMonth[3][0]
for DF in DFListMonth[3][1:]:
    CongDF_Sep17 = CongDF_Sep17.append(DF)
# October
CongDF_Oct17 = DFListMonth[4][0]
for DF in DFListMonth[4][1:]:
    CongDF_Oct17 = CongDF_Oct17.append(DF)
# November
CongDF_Nov17 = DFListMonth[5][0]
for DF in DFListMonth[5][1:]:
    CongDF_Nov17 = CongDF_Nov17.append(DF)
# December
CongDF_Dec17 = DFListMonth[6][0]
for DF in DFListMonth[6][1:]:
    CongDF_Dec17 = CongDF_Dec17.append(DF)
```

```python
CongDF_June17.to_csv("s3://502finalprojbucky/congresstweets/data/June2017.csv",index=False)
CongDF_July17.to_csv("s3://502finalprojbucky/congresstweets/data/July2017.csv",index=False)
CongDF_Aug17.to_csv("s3://502finalprojbucky/congresstweets/data/Aug2017.csv",index=False)
CongDF_Sep17.to_csv("s3://502finalprojbucky/congresstweets/data/Sep2017.csv",index=False)
CongDF_Oct17.to_csv("s3://502finalprojbucky/congresstweets/data/Oct2017.csv",index=False)
CongDF_Nov17.to_csv("s3://502finalprojbucky/congresstweets/data/Nov2017.csv",index=False)
CongDF_Dec17.to_csv("s3://502finalprojbucky/congresstweets/data/Dec2017.csv",index=False)
```
